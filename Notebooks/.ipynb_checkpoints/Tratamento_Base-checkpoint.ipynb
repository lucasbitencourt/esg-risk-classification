#### Leitura da base de dados do S&P 500 diretamento do Yahoo Finance utilizando a biblioteca yfinance, que em seguida será exportado para CSV

import pandas as pd
import yfinance as yf
import time

# 1. Captura a lista de empresas do S&P 500 da Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
tabela = pd.read_html(url)
sp500_df = tabela[0]
tickers = sp500_df['Symbol'].tolist()

# Corrige tickers com ponto
tickers = [ticker.replace('.', '-') for ticker in tickers]

# 2. Coleta todas as informações disponíveis de cada empresa
def get_all_info(ticker):
    try:
        info = yf.Ticker(ticker).info
        info['Ticker'] = ticker  # adiciona a identificação
        return info
    except Exception as e:
        print(f"Erro com {ticker}: {e}")
        return None

# 3. Loop para coletar os dados
dados_completos = []
for ticker in tickers:
    print(f"Coletando: {ticker}")
    data = get_all_info(ticker)
    if data:
        dados_completos.append(data)
    time.sleep(1)  # pausa para evitar bloqueio da API

# 4. Criar DataFrame com todas as colunas possíveis
df_completo = pd.DataFrame(dados_completos)

# 5. Exportar
df_completo.to_excel('sp500_dados_completos.xlsx', index=False)
df_completo.to_csv('sp500_dados_completos.csv', index=False)

print("Coleta completa finalizada.")

#### Preenchendo valores nulos com 0 nas colunas indicadas

```python
# Lista das colunas a preencher
colunas_para_preencher = [
    'fiveyearavgdividendyield',
    'dividendrate',
    'dividendyield',
    'lastdividendvalue'
]

# Preenche os valores nulos com 0 nessas colunas
df.loc[:, colunas_para_preencher] = df[colunas_para_preencher].fillna(0)
```

#### Preenchendo valores nulos com média e moda

```python
import pandas as pd

# Dicionário com as colunas e o método de preenchimento
colunas_preenchimento = {
    'trailingpegratio': 'media',
    'earningsgrowth': 'media',
    'earningsquarterlygrowth': 'media',
    'debttoequity': 'media',
    'enterprisetoebitda': 'media',
    'returnonequity': 'media',
    'ebitda': 'media',
    'freecashflow': 'media',
    'trailingpe': 'media',
    'quickratio': 'media',
    'currentratio': 'media',
    'recommendationmean': 'media',
    'enterprisetorevenue': 'media',
    'enterprisevalue': 'media',
    'totaldebt': 'media',
    'totalcashpershare': 'media',
    'totalcash': 'media',
    'overallrisk': 'media',
    'priceepscurrentyear': 'media',
    'epscurrentyear': 'media',
    'beta': 'media',
    'auditrisk': 'moda',
    'compensationrisk': 'moda',
    'shareholderrightsrisk': 'moda',
    'boardrisk': 'moda',
    'shortpercentoffloat': 'media',
    'numberofanalystopinions': 'moda',
    'revenuegrowth': 'media',
    'operatingcashflow': 'media',
    'epsforward': 'media',
    'forwardeps': 'media',
    'targetmedianprice': 'media',
    'fulltimeemployees': 'media',
    'postmarketchangepercent': 'media',
    'postmarketchange': 'media'
    
}

# Aplica o preenchimento apropriado
for coluna, metodo in colunas_preenchimento.items():
    if metodo == 'media':
        valor = pd.to_numeric(df[coluna], errors='coerce').mean(skipna=True)
    elif metodo == 'moda':
        valor = df[coluna].mode(dropna=True)
        valor = valor.iloc[0] if not valor.empty else None
    else:
        continue  # ignora métodos não reconhecidos

    # Preenche os valores nulos com o valor calculado
    df[coluna] = df[coluna].fillna(valor)
```

#### Tratando colunas com range de valores ('37.59 - 141.63')

```python
# Trata regularmarketdayrange
df[['regularmarketdayrange_min', 'regularmarketdayrange_max']] = (
    df['regularmarketdayrange']
    .str.split(' - ', expand=True)
    .astype(float)
)

# Trata fiftytwoweekrange
df[['fiftytwoweekrange_min', 'fiftytwoweekrange_max']] = (
    df['fiftytwoweekrange']
    .str.split(' - ', expand=True)
    .astype(float)
)

# Dropando colunas originais
df = df.drop(columns=['regularmarketdayrange', 'fiftytwoweekrange'])
```
